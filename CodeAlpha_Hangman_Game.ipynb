```python
import random

words = ["table","aissms","code","laptop","game","codealpha"]  # list Concepts used
word = random.choice(words) #random Concepts used
guessed = ["_"] * len(word)
attempts = 6
used_letters = []

print("Welcome to hangman! You have 6 trys")

# while Concepts used
while attempts > 0 and "_" in guessed:
    print("Word so far:", " ".join(guessed))
    
    # it's show you used letters 
    if used_letters:
        print("used letters:", " ".join(used_letters))
    
    guess = input("guess a letter: ").lower()
    
    # check if its a letter and only one
    if len(guess) != 1 or not guess.isalpha():
        print("enter only ONE letter!")
        continue
    
    # it's check if you already use this letter
    if guess in used_letters:
        print("you already used this!")
        continue
    
    used_letters.append(guess)
    
    # it's check if letter is in thw word then it's show good guess
    # if else Concepts used
    if guess in word:
        print("good guess!")
        for i in range(len(word)):
            if word[i] == guess:
                guessed[i] = guess
    else:
        attempts = attempts - 1
        print("wrong guess! tries left:", attempts)

# it's check you win or loss you game
if "_" not in guessed:
    print("You won! The word is:", word)
else:
    print("You loss! The word is:", word)
    print("Better luck next time.")
```
